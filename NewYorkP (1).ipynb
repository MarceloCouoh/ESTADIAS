```python
import requests
```

```python
page = requests.get("https://nypost.com/")
```

```python
page
```

```python
page.status_code
```

```python
pagess=(page.content.strip())
```

```python
print(pagess)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify())
```

```python
panel= soup.find(class_="background-wrapper")
```

```python
forecast_items = panel(class_="headline-wrapper")
```

```python
tonight = forecast_items[0]
```

```python
print(tonight)
```

```python
period = tonight.find(class_="headline-container").get_text()
```

```python
print(period)
```

```python
period_tags = panel.select(".headline-wrapper .headline-container")
```

```python
periods = [pt.get_text() for pt in period_tags]
```

```python
periods
```

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "noticias": periods,
   
})
weather
```
