```python
import requests
```

```python
page=requests.get("https://www.newsday.com/")
```

```python
page
```

```python
pagess =(page.content.strip())
```

```python
print(pagess)
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(page.content, 'html.parser')
```

```python
print(soup.prettify)
```

```python
panel= soup.find(class_="top-stories")
```

```python
forecast_items = panel(class_="top-stories-cont")
```

```python
tonight = forecast_items[0]
```

```python
print(tonight)
```

```python
period = tonight.find(class_="item").get_text()
```

```python
print(period)
```

```python
period_tags = panel.select(".top-stories-cont .item")
```

```python
periods = [pt.get_text() for pt in period_tags]
```

```python
periods
```

```python
import pandas as pd
```

```python
weather = pd.DataFrame({
    "noticias": periods,
   
})
weather
```
